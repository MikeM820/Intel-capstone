```python
def calculate_duration_in_each_hour(start_hour, end_hour):
    secs_in_each_hour_24 = {h:0 for h in range(num_hours_in_a_day)}
    if end_hour > start_hour:
        for curr_hour in range(start_hour, end_hour+1):
            if start_hour == curr_hour:
                secs_in_each_hour_24[curr_hour] = int(
                    (hour_end_time[curr_hour] - start_time).total_seconds()
                )
            elif start_hour < curr_hour < end_hour:
                secs_in_each_hour_24[curr_hour] = int(
                    (hour_end_time[curr_hour] - hour_end_time[curr_hour-1]).total_seconds()
                )
            elif curr_hour == end_hour:
                secs_in_each_hour_24[curr_hour] = int(
                    (end_time - hour_end_time[curr_hour-1]).total_seconds() + 1
                )
    else:
        secs_in_each_hour_24[start_hour] = (end_time - start_time).total_seconds()
    return secs_in_each_hour_24

    
```
